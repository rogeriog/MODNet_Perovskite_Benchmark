### I will try to use bagging with linear regression to obtain some OFM mixed features that better reproduce the target of perovskite energies

```python
from modnet.preprocessing import MODData
dataOFM=MODData.load('DATAFILES/matbench_perovskites_moddataOFM.pkl.gz')
```

```python
## OFM feats only
OFMfeats=dataOFM.df_featurized.filter(like="OFM")
OFMfeats
```

```python
import numpy as np
# from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.ensemble import BaggingRegressor
X=OFMfeats
y=dataOFM.df_targets.values
regr = BaggingRegressor(base_estimator=Ridge(positive=True), n_estimators=10, random_state=0).fit(X, y)
```

```python
regr.estimators_[0].coef_
```
